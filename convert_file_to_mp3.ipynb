```python
from pathlib import Path
from tqdm import tqdm
from src.mp3_converter import MP3Converter
%load_ext autoreload
%autoreload 2
```

```python
def filter_filepaths(filepaths):
    filepaths =  [filepath for filepath in filepaths if 'REC' in str(filepath)]
    return filepaths
```

```python
DIR = Path().absolute().parent.parent.parent.joinpath('Music/Soulseek Songs')
music_file_extensions = ['m4a','opus', 'wav', 'flac', 'aiff']
filepaths = [elem for elem in list(DIR.rglob('*')) if elem.suffix[1:] in music_file_extensions]
```

```python
for filepath in tqdm(filepaths):
    file_extension = filepath.suffix
    conv = MP3Converter(filepath, file_extension, "mp3")
    conv.execute()
# conv.make_dir()
# print("WAV's converted to MP3's and stored in new directory!")
```
